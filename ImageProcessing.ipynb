**Image Processing with Python**
  
    1.Reading an Image and converting into a numpy Array
    2.Resizing and Image
    3.RGB to GrayScale Image

**Libraries Req:**

    1.Matplotlib.image
    2.Pillow
    3.OpenCV(cv2)

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

Reading the Image

```python
img=mpimg.imread('img1.jpg')
type(img)
print(img.shape)
```

```python
#display the image from numpy array
img_plot=plt.imshow(img)
```

Resizing the Image using Pillow Library

```python
from PIL import Image
```

```python
img=Image.open('img1.jpg').resize((200,200))
```

```python
img.save('img1_resized.jpg')
```

```python
img_plot=plt.imshow(img)
```

Converting RGB Images into GrayScale Images

```python
#Importing OpenCV Libraries
import cv2
```

```python
img=cv2.imread('img1_resized.jpg')
```

```python
gray_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) 
print(gray_image.shape)
```
